```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import math
import statistics

file = "test.csv" 
data = pd.read_csv(file) 
df = pd.DataFrame(data)

print(df)

total_row = df.shape[0]

i=0
tot_weight = [0,0,0,0,0,0]
tot = [0,0,0,0,0,0]

for i in range (total_row):
    if (int(df["min"][i])<10):
        tot[0] += float(df["1st m $"][i]) * int((df["1st m #"][i]))
        tot_weight[0] += int((df["1st m #"][i]))
    elif (int(df["min"][i])<20):
        tot[1] += float(df["1st m $"][i]) * int((df["1st m #"][i]))
        tot_weight[1] += int((df["1st m #"][i]))
    elif (int(df["min"][i])<30):
        tot[2] += float(df["1st m $"][i]) * int((df["1st m #"][i]))
        tot_weight[2] += int((df["1st m #"][i]))
    elif (int(df["min"][i])<40):
        tot[3] += float(df["1st m $"][i]) * int((df["1st m #"][i]))
        tot_weight[3] += int((df["1st m #"][i]))
    elif (int(df["min"][i])<50):
        tot[4] += float(df["1st m $"][i]) * int((df["1st m #"][i]))
        tot_weight[4] += int((df["1st m #"][i]))
    elif (int(df["min"][i])<60):
        tot[5] += float(df["1st m $"][i]) * int((df["1st m #"][i]))
        tot_weight[5] += int((df["1st m #"][i]))

tmp = [0,0,0,0,0,0]
tmp2 = [0,0,0,0,0]

for i in range(6):
    tmp[i] = math.log(tot[i]/tot_weight[i])

for i in range(5):
    tmp2[i] = tmp[i+1] - tmp[i]
    
final = statistics.stdev(tmp2[0:4])

print(tmp)   
print(tmp2)
print(final)

#df = df.insert(6,"ln(price)", tmp)
#print(df)
#df["ln(price)"] = 
#print(df)
#sort = df["prod_id"].value_counts(ascending = True)
#print(sort)
#print(tmp2)
#print(statistics.stdev(tmp2[4:0]))
```
